```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
image = cv2.imread("image.jpg",1)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
blurred = cv2.GaussianBlur(gray,(5,5),0)
```

```python
canny = cv2.Canny(blurred,0,100)
(cnts,_) = cv2.findContours(canny.copy(),cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
coints = image.copy()
frame = image.copy()
cv2.drawContours(frame,cnts,-1,(0,255,0),1)
plt.imshow(frame);
```

```python
r = np.full((gray.shape[0], image.shape[1],1),0)
g = np.full((gray.shape[0], image.shape[1],1),0)
b = np.full((gray.shape[0], image.shape[1],1),0)
arr = np.dstack((r,g,b))
```

```python
frame = image.copy()
for cnt in cnts:
    x ,y,w,h = cv2.boundingRect(cnt)
    cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),1)
plt.imshow(frame);
```
